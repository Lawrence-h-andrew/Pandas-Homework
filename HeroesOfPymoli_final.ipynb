```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
#purchase_data = purchase_data.set_index('SN')
purchase_data.head(10)
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
uniqueitems = purchase_data['Item ID'].nunique()
print("Unique Items: ", uniqueitems)
averageprice = purchase_data['Price'].mean()
print("Average Price: ", averageprice)
totalnumberofpurchases = purchase_data['Purchase ID'].count()
print("Total number of Purchases: ", totalnumberofpurchases)
totalrevenue = purchase_data['Price'].sum()
print("Total Revenue: ", totalrevenue)
Gendersummary = pd.DataFrame({"Unique Items: ": uniqueitems,
                                 "Average Price: ": [averageprice],
                                 "Total number of Purchases: ": [totalnumberofpurchases],
                                 "Total Revenue: ": [totalrevenue]})

Gendersummary["Average Price: "]=Gendersummary["Average Price: "].astype(float).map("${:,.2f}".format)
Gendersummary["Total Revenue: "]=Gendersummary["Total Revenue: "].astype(float).map("${:,.2f}".format)

Gendersummary
```

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results

* Display the summary data frame

```python
malepurchasecount = purchase_data.query('Gender == "Male"').Gender.count() 
femalepurchasecount = purchase_data.query('Gender == "Female"').Gender.count()
otherpurchasecount = purchase_data.query('Gender == "Other / Non-Disclosed"').Gender.count()
#print(malepurchasecount)
malepurchaseavg = purchase_data.query('Gender == "Male"').Price.mean()
femalepurchaseavg = purchase_data.query('Gender == "Female"').Price.mean()
otherpurchaseavg = purchase_data.query('Gender == "Other / Non-Disclosed"').Price.mean()

maletotalpurchase = purchase_data.query('Gender == "Male"').Price.sum()
femaletotalpurchase = purchase_data.query('Gender == "Female"').Price.sum()
othertotalpurchase = purchase_data.query('Gender == "Other / Non-Disclosed"').Price.sum()

avgtotalmale = maletotalpurchase / malecount
avgtotalfemale = femaletotalpurchase / femalecount
avgtotalother = othertotalpurchase / othercount

summarytable = pd.DataFrame({"Purchase Count": [malepurchasecount, femalepurchasecount, otherpurchasecount],
                             "Avg Purchase Price": [malepurchaseavg, femalepurchaseavg, otherpurchaseavg],
                             "Total Purchase": [maletotalpurchase, femaletotalpurchase, othertotalpurchase],
                             "Avg Total Purchase": [avgtotalmale, avgtotalfemale, avgtotalother]},
                              index= ['Male', 'Female', 'Others'])
summarytable["Avg Purchase Price"] = summarytable["Avg Purchase Price"].astype(float).map("${:,.2f}".format)
summarytable["Total Purchase"] = summarytable["Total Purchase"].astype(float).map("${:,.2f}".format)
summarytable["Avg Total Purchase"] = summarytable["Avg Total Purchase"].astype(float).map("${:,.2f}".format)

summarytable

#after review I could have indexed gender, .loc ["Male", "Price"]
```
